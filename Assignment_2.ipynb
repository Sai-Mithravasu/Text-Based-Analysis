# Importing Regex

```python
import regex as re 
```

```python
#Defining a Function of the regex 
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

# Downloading StopWords

```python
# Importing all the stop words
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
# Defining a function to remove stop words
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
```

# Creating a Pipeline

```python
#  Creating a pipeline
pipeline = [str.lower, tokenize, remove_stop]
```

```python
# Defining a function to run the pipeline 
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Applying a Pipe line

```python
# Applying he function to text 
df['tokens_d'] = df['description'].apply(prepare, pipeline = pipeline)
```

```python
 # Creating a new column for the number of tokens
df['num_tokens_d'] = df['description'].map(len)
df.tail()
```

# Importing Counter

```python
#Importing a function Counter
from collections import Counter
```

```python
# Printing the 5 most occurred tokens
counter_d = Counter()
df['tokens_d'].map(counter_d.update)
print(counter_d.most_common(5))
```

```python
freq_dfd = pd.DataFrame.from_dict(counter_d,orient='index', columns=['freq'])
freq_dfd = freq_dfd.query('freq >= 1')
freq_dfd.index.name = 'token'
```

# Most Frequent Words

```python
freq_dfd = freq_dfd.sort_values('freq', ascending=False)
freq_dfd.head(10)
```

```python
freq_dfd.describe()
```

# Visualisation of Number of Tokens and Most frequent Words

```python
# Step 2: Create a Histogram for 'num_tokens_d'
plt.figure(figsize=(8, 6))
plt.hist(df['num_tokens_d'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution of Number of Tokens in Descriptions')
plt.xlabel('Number of Tokens')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Plotting a inverted bar graph for most frequent words 
ax = freq_dfd.head(15).plot(kind='barh', width=0.95, figsize=(10,6),edgecolor ='white')
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
